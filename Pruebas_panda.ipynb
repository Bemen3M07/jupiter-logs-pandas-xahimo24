```python
import pandas as pd
notas = pd.DataFrame({'Profesor': ['Emili','Isabel','Jose','Emili','Isabel'],
                    'Alumno':['Xavier','Juan','Carlos','Maradona','Paquito'],
                    'M01':[8,6,5,5,5],
                    'M02':[6,4,8,4,9],
                    'M03':[7,5,7,3,8],
                    'M04':[9,7,4,8,9],
                    'M05':[4,5,7,8,4]})
notas
```

```python
notas[["M01", "M02", "M03", "M04", "M05"]].mean()
```

```python
notas.groupby('Alumno', as_index=False)[["M01", "M02", "M03", "M04", "M05"]].mean()
```

```python
import matplotlib.pyplot as plt
notas.plot(x="Alumno", y=["M01", "M02"])
plt.rcParams["figure.figsize"] = (10,5)
plt.show()
```

```python
notas[["M01", "M02", "M03", "M04", "M05"]].mean(axis = 1)
```
